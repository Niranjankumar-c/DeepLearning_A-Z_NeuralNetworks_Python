# Artificial Neural Network

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
dataset.Exited.value_counts()
```

```python
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
```

```python
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
```

```python
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
```

```python
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Keras

```python
import warnings
with warnings.catch_warnings():
    warnings.filterwarnings("ignore",category=FutureWarning)
```

```python
import keras
```

```python
# Fitting classifier to the Training set
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initilizing the ANN

classifier = Sequential()
```

```python
#adding the input layer and first hidden layer

classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu", input_dim = 11))
```

```python
#adding the second hidden layer

classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu"))
```

```python
#adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = "uniform", activation = "sigmoid"))

#compling the ANN
classifier.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

### K Fold Cross Validation

```python
#import the keras wrapper to combine kfold and keras
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu", input_dim = 11))
    classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu"))
    classifier.add(Dense(units=1, kernel_initializer="uniform", activation="sigmoid"))
    classifier.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    
    return(classifier)
```

```python
newclassifier = KerasClassifier(build_fn=build_classifier, batch_size = 10, epochs = 100)
```

```python
accuracies = cross_val_score(estimator=newclassifier, X=X_train, y=y_train,cv = 10,verbose=2)
```

```python
accuracies.mean()
```

```python
accuracies
```

```python
accuracies.std()
```

### Dropout Regularization & Parameter Tunning

```python
# Fitting classifier to the Training set
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#Initilizing the ANN

classifier = Sequential()
```

```python
#adding the input layer and first hidden layer with dropput

classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu", input_dim = 11))
classifier.add(Dropout(rate=0.1))
```

```python
#adding the second hidden layer

classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu"))
classifier.add(Dropout(rate = 0.1))
```

```python
#adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = "uniform", activation = "sigmoid"))

#compling the ANN
classifier.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

## Tunning

```python
#import the keras wrapper to combine kfold and keras
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

classifier = KerasClassifier(build_fn = build_classifier)
parameters = {'batch_size': [25, 32],
              'epochs': [100, 500],
              'optimizer': ['adam', 'rmsprop']}
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 5, verbose = 2)
```

```python
grid_search = grid_search.fit(X_train, y_train)
```

```python
best_parameters = grid_search.best_params_
```

```python
best_parameters
```

```python
best_accuracy = grid_search.best_score_
```
